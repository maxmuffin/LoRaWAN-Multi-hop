```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Test1, Test2, Test3_500ms, Test3_5sec, Test3_10sec, AllTest3, AllTests, AllTest
# AllTests contiene tutte le rilevazioni per Test3 in modalit√† di confronto
# AllTest confronta solo la media dei Test3 con gli altri Test

pageName = 'AllTest'
pageTitle = pageName + ' - Latency'
saveGraphs = False

savePath = 'images/graphs/latency/'+pageName
df = pd.read_excel (r'data/latency_nuvolaPunti.xlsx', sheet_name=pageName)

delayTicks = [300, 500, 750, 1000, 1500, 2000] 
delayPoints = df['delay'].to_numpy()
latencyPoints = df['latency'].to_numpy()
marginErrPoints = df['marginErr'].to_numpy()
maxPoints = df['max'].to_numpy()
minPoints = df['min'].to_numpy()

plt.style.use('seaborn-whitegrid')
paper_rc = {'lines.linewidth': 1, 'lines.markersize': 10}    
sns.set_context("paper", rc = paper_rc)  
sns.set(style="ticks", palette="muted")

ticksSize = 20
fontSize = 20
legendTitle = 30
markerLegend = 15


#fig, ax = plt.subplots()
```

```python
if pageName in ["AllTests", "AllTest3", "AllTest"]:
    plt.figure(figsize=(20,10))
    #df.plot.scatter('delay', 'latency', color='red', s=0.5)
    sns.scatterplot(x="delay", y="latency", hue="Test", palette="muted",data=df)
    #plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., fontsize=fontSize)
    plt.legend(fontsize=fontSize, title_fontsize=legendTitle, markerscale=markerLegend/10, loc='best')
    plt.xticks(delayTicks, fontsize=ticksSize)
    plt.yticks(fontsize =ticksSize)
    plt.xlabel("Time between pkt sent (ms)", fontsize=fontSize)
    plt.ylabel("Latency (s)", fontsize=fontSize)
    if saveGraphs == True:
        plt.savefig(savePath+'_scatterPlot.png', format='png', dpi=300,  bbox_inches='tight', pad_inches=0.2)
else:
    plt.figure(figsize=(20,10))
    #df.plot.scatter('delay', 'latency', color='red', s=0.5)
    sns.scatterplot(x="delay", y="latency",data=df)
    plt.xticks(delayTicks, fontsize=ticksSize)
    plt.yticks(fontsize =ticksSize)
    plt.xlabel("Time between pkt sent (ms)", fontsize=fontSize)
    plt.ylabel("Latency (s)", fontsize=fontSize)
    if saveGraphs == True:
        plt.savefig(savePath+'_scatterPlot.png', format='png', dpi=300,  bbox_inches='tight', pad_inches=0.2)
```

```python
if pageName not in ["AllTests", "AllTest3", "AllTest"]:
    #Scatter plot with error bar
    plt.figure(figsize=(20,10))
    plt.errorbar(delayPoints, latencyPoints,  yerr=marginErrPoints, markersize=3, fmt='o', color ='red', ecolor='black', capsize=2, barsabove = False, alpha =0.1)
    #plt.title("Scatter plot with error bars - "+pageTitle, fontsize=fontSize)
    plt.xticks(delayTicks, fontsize = ticksSize)
    plt.yticks(fontsize = ticksSize)
    plt.xlabel("Time between pkt sent (ms)", fontsize=fontSize)
    plt.ylabel("Latency (s)", fontsize=fontSize)

    if saveGraphs == True:
        plt.savefig(savePath+'_dispersion.png', format='png', dpi=300,  bbox_inches='tight', pad_inches=0.2)
    
```

```python
if pageName in ["AllTests", "AllTest3", "AllTest"]:
    #Error interval mean plot
    plt.figure(figsize=(20,10))
    sns.pointplot('delay', 'latency', hue='Test', markers=["2", "x", "+", "^","1"], data=df, capsize=.05, yerr= marginErrPoints, height=5, aspect = 2, err_style="bars", ci=90)
    #plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., fontsize=fontSize)
    plt.legend(fontsize=fontSize, title_fontsize=legendTitle, markerscale=markerLegend/10)
    #plt.title("Error plot - "+pageTitle, fontsize=fontSize)
    plt.xlabel("Time between pkt sent (ms)", fontsize=fontSize)
    plt.ylabel("Latency (s)", fontsize=fontSize)
    plt.xticks(fontsize=ticksSize)
    plt.yticks(fontsize =ticksSize)
    if saveGraphs == True:
        plt.savefig(savePath+'_linePlotWithErrorBar.png', format='png', dpi=300,  bbox_inches='tight', pad_inches=0.2)
else:
    #Error plot
    plt.figure(figsize=(20,10))
    sns.pointplot('delay', 'latency', data=df, capsize=.05, yerr= marginErrPoints,  err_style="bars", ci=90)
    #plt.title("Error plot - "+pageTitle, fontsize=fontSize)
    plt.xlabel("Time between pkt sent (ms)", fontsize=fontSize)
    plt.ylabel("Latency (s)", fontsize=fontSize)
    plt.xticks(fontsize=ticksSize)
    plt.yticks(fontsize =ticksSize)
    if saveGraphs == True:
        plt.savefig(savePath+'_linePlotWithErrorBar.png', format='png', dpi=300,  bbox_inches='tight', pad_inches=0.2)
```

```python
if pageName in ["AllTests", "AllTest3", "AllTest"]:
    #Smote plot
    plt.figure(figsize=(20,10))
    ax = sns.lmplot('delay', 'latency', hue ='Test', legend_out=False, data=df,  ci=None, order=4,height=10, aspect = 2, truncate=True, scatter_kws={"s": 0.5})
    #plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    #plt.title("Smoted line plot - "+pageTitle, fontsize=fontSize)
    plt.xticks(delayTicks, fontsize=ticksSize)
    plt.yticks(fontsize =ticksSize)
    plt.xlabel("Time between pkt sent (ms)", fontsize=fontSize)
    plt.ylabel("Latency (s)", fontsize=fontSize)
    plt.legend(fontsize=fontSize, title_fontsize=legendTitle, markerscale=markerLegend)
    
    if saveGraphs == True:
        plt.savefig(savePath+'_smoteInterpolation.png', format='png', dpi=300,  bbox_inches='tight', pad_inches=0.2)
        
    plt.figure(figsize=(20,10))
    ax = sns.lmplot('delay', 'latency', hue ='Test', legend_out=False, data=df,  ci=90, order=4,height=10, aspect = 2, truncate=True, scatter_kws={"s": 0.5})
    #plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    #plt.title("Smoted line plot - "+pageTitle, fontsize=fontSize)
    plt.xticks(delayTicks, fontsize=ticksSize)
    plt.yticks(fontsize =ticksSize)
    plt.xlabel("Time between pkt sent (ms)", fontsize=fontSize)
    plt.ylabel("Latency (s)", fontsize=fontSize)
    plt.legend(fontsize=fontSize, title_fontsize=legendTitle, markerscale=markerLegend)
    
    if saveGraphs == True:
        plt.savefig(savePath+'_smoteInterpolationWithCI.png', format='png', dpi=300, bbox_inches='tight', pad_inches=0.2)
else:
    #Smote plot
    plt.figure(figsize=(20,10))
    g = sns.lmplot('delay', 'latency',data=df,  ci=90, order=4, height=10, aspect = 2, truncate=True, legend_out=True, scatter_kws={"s": 5})
    #plt.title("Smoted line plot with CI - "+pageTitle, fontsize=fontSize)
    plt.xticks(delayTicks, fontsize=ticksSize)
    plt.yticks(fontsize =ticksSize)
    
    plt.xlabel("Time between pkt sent (ms)", fontsize=fontSize)
    plt.ylabel("Latency (s)", fontsize=fontSize)
    
    if pageName == "Test1":
        g.set(ylim=(0.25, 0.41))
    if pageName == "Test2":
        g.set(ylim=(0.7, 1.1))
    if pageName == "Test3 500ms":
        g.set(ylim=(17, 20.5))
    if pageName == "Test3 5sec":
        g.set(ylim=(22, 25))
    if pageName == "Test3 10sec":
         g.set(ylim=(21, 31))
    if saveGraphs == True:
        plt.savefig(savePath+'_smoteinterpolationWithCI.png', format='png', dpi=300, bbox_inches='tight', pad_inches=0.2, transparent=False)
    
    plt.figure(figsize=(20,10))
    g = sns.lmplot('delay', 'latency',data=df,  ci=None, order=4, height=10, aspect = 2, truncate=True, legend_out=True, scatter_kws={"s": 5})
    #plt.title("Smoted line plot - "+pageTitle, fontsize=fontSize)
    plt.xticks(delayTicks, fontsize=ticksSize)
    plt.yticks(fontsize =ticksSize)
    plt.xlabel("Time between pkt sent (ms)", fontsize=fontSize)
    plt.ylabel("Latency (s)", fontsize=fontSize)
    
    if pageName == "Test1":
        g.set(ylim=(0.25, 0.41))
    if pageName == "Test2":
        g.set(ylim=(0.7, 1.1))
    if pageName == "Test3 500ms":
        g.set(ylim=(17, 20.5))
    if pageName == "Test3 5sec":
        g.set(ylim=(22, 25))
    if pageName == "Test3 10sec":
         g.set(ylim=(21, 31))
    if saveGraphs == True:
        plt.savefig(savePath+'_smoteinterpolation.png', format='png', dpi=300, bbox_inches='tight', pad_inches=0.2)
```

```python
if pageName in ["AllTests", "AllTest3", "AllTest"]:
    #Smote plot - confronto
    plt.figure(figsize=(20,10))
    sns.set(style="ticks", palette="muted", font_scale = 2)
    g = sns.lmplot(x="delay", y="latency", col="Test", hue="Test", data=df,
           col_wrap=2, ci=90, palette="muted", height=5, aspect = 2, order=4, truncate = True,
           scatter_kws={"s": 4, "alpha": 1})
    g = (g.set_axis_labels("Time between pkt sent (ms)", "Latency (s)"))
    plt.xticks(delayTicks)
    
    if saveGraphs == True:
        plt.savefig(savePath+'_confrontoSmootedLine.png', format='png', dpi=300,  bbox_inches='tight', pad_inches=0.2)
    
    plt.figure(figsize=(20,10))
    sns.set(style="ticks", palette="muted", font_scale = 1.5)
    g = sns.lmplot('delay', 'latency', col="Test", hue="Test", data=df, scatter_kws={"s": 1, "alpha": 1},
               palette="muted", y_jitter=.02, logistic=False, truncate=True)
    g = (g.set_axis_labels("Time between pkt sent (ms)", "Latency (s)"))
    
    g.set(ylim=(0, 32))
    plt.xticks(delayTicks)
    
    if saveGraphs == True:
        plt.savefig(savePath+'_confrontoRegressionLine.png', format='png', dpi=300,  bbox_inches='tight', pad_inches=0.2)
```

```python
if pageName in ["AllTests", "AllTest3", "AllTest"]:
    plt.figure(figsize=(20,10))
    
    sns.set(style="ticks", palette="muted", font_scale = 2)
    sns.boxplot('delay', 'latency', palette="muted", data=df, hue="Test", width=0.3, dodge = False) #notch = True
    #plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., fontsize=fontSize)
    plt.legend(fontsize=fontSize, title_fontsize=legendTitle, markerscale=markerLegend,loc='best')
    #plt.title("Box plot - "+pageTitle, fontsize=fontSize)
    plt.xlabel("Time between pkt sent (ms)", fontsize=fontSize)
    plt.ylabel("Latency (s)", fontsize=fontSize)
    plt.xticks(fontsize=ticksSize)
    plt.yticks(fontsize =ticksSize)
    sns.despine(offset=10, trim=True)
    
    if saveGraphs == True:
        plt.savefig(savePath+'_boxPlot.png', format='png', dpi=300,  bbox_inches='tight', pad_inches=0.2) 
else:
    plt.figure(figsize=(20,10))
    sns.set(style="ticks", palette="muted", font_scale = 2)
    sns.boxplot('delay', 'latency', palette="muted", data=df,width=0.3, dodge = False)
    #sns.swarmplot('delay', 'latency', data=df, color=".25")
    #plt.title("Box plot - "+pageTitle, fontsize=fontSize)
    plt.xlabel("Time between pkt sent (ms)", fontsize=fontSize)
    plt.ylabel("Latency (s)", fontsize=fontSize)
    #plt.xticks(fontsize=ticksSize)
    #plt.yticks(fontsize =ticksSize)
    
    sns.despine(offset=10, trim=True)

    if saveGraphs == True:
        plt.savefig(savePath+'_boxPlot.png', format='png', dpi=300,  bbox_inches='tight', pad_inches=0.2) 
```
